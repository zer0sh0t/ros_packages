```python
import bagpy
import pandas as pd
from bagpy import bagreader
import matplotlib.pyplot as plt
```

```python
b = bagreader('puck/test_run.bag')
gps_data = b.message_by_topic('/gps')
df_gps = pd.read_csv(gps_data)
```

```python
df_gps
```

```python
def plot_graph(xax, yax, xax_name=None, yax_name=None):
    plt.plot(xax, yax)
    if xax_name != None: 
        plt.xlabel(xax_name)
    if yax_name != None:
        plt.ylabel(yax_name)
    plt.show()
```

```python
plot_graph(df_gps['Time'].tolist(), df_gps['Latitude'].tolist(), 'time', 'latitude')
```

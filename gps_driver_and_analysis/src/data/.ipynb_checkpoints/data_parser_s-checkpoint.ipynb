```python
import bagpy
import numpy as np
import pandas as pd
import seaborn as sns
from bagpy import bagreader
import matplotlib.pyplot as plt
```

```python
b = bagreader('stationary_bag.bag')
gps_data = b.message_by_topic('/gps')
df_gps = pd.read_csv(gps_data)
```

```python
df_gps
```

```python
df_gps.to_csv('stationary_gps.csv')
```

```python
df_gps['Time'] = df_gps['Time'] - min(df_gps['Time'])
```

```python
df_gps['UTM_easting'] = df_gps['UTM_easting'] - min(df_gps['UTM_easting'])
```

```python
df_gps['UTM_northing'] = df_gps['UTM_northing'] - min(df_gps['UTM_northing'])
```

```python
df_gps
```

```python
def plot_graph(xax, yax):
    plt.plot(df_gps[xax].tolist(), df_gps[yax].tolist())
    plt.xlabel(xax)
    plt.ylabel(yax)
    plt.title(f'{yax} vs {xax}')
    plt.show()
```

```python
plot_graph('Time', 'Latitude')
```

```python
plot_graph('Time', 'Longitude')
```

```python
plot_graph('Time', 'Altitude')
```

```python
plot_graph('Time', 'UTM_easting')
```

```python
plot_graph('Time', 'UTM_northing')
```

```python
plot_graph('Time', 'Zone')
```

```python
sns.set_style('white')
sns.set_context("paper", font_scale = 2)
sns.displot(data=df_gps, x="UTM_easting", kind="hist", bins=100, aspect=1.5)
```

```python
sns.displot(data=df_gps, x="UTM_northing", kind="hist", bins=100, aspect=1.5)
```

```python
sns.displot(data=df_gps, x="Altitude", kind="hist", bins=100, aspect=1.5)
```

```python
def get_bounds(col_name):
    col = df_gps[col_name].tolist()
    mean = np.mean(col)
    ub = max(col) - mean
    lb = min(col) - mean
    print(f'upper bound: {ub}')
    print(f'mean: {mean}')
    print(f'lower bound: {lb}')
```

```python
get_bounds('UTM_easting')
```

```python
get_bounds('UTM_northing')
```

```python
get_bounds('Altitude')
```

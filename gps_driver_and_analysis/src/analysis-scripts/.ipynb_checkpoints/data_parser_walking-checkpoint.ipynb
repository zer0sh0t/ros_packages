```python
import bagpy
import numpy as np
import pandas as pd
import seaborn as sns
from bagpy import bagreader
import matplotlib.pyplot as plt
```

```python
b = bagreader('../data/walking_data.bag')
gps_data = b.message_by_topic('/gps')
df_gps = pd.read_csv(gps_data)
```

```python
df_gps
```

```python
df_gps['Time'] = df_gps['Time'] - min(df_gps['Time'])
```

```python
df_gps['UTM_easting'] = df_gps['UTM_easting'] - np.mean(df_gps['UTM_easting'])
```

```python
df_gps['UTM_northing'] = df_gps['UTM_northing'] - np.mean(df_gps['UTM_northing'])
```

```python
df_gps
```

```python
def plot_graph(xax, yax, scatter=False):
    if scatter:
        plt.scatter(df_gps[xax].tolist(), df_gps[yax].tolist())
    else:
        plt.plot(df_gps[xax].tolist(), df_gps[yax].tolist())
    plt.xlabel(xax)
    plt.ylabel(yax)
    plt.title(f'{yax} vs {xax}')
    plt.show()
```

```python
plot_graph('UTM_easting', 'UTM_northing')
```

```python
plot_graph('Time', 'Altitude')
```

```python
x = df_gps['UTM_northing'].tolist()
y = df_gps['UTM_easting'].tolist()
m, b = np.polyfit(x, y, 1)

def best_line(x_):
    return m * np.array(x_) + b

y_ideal = best_line(x)
```

```python
plt.plot(df_gps['UTM_northing'].tolist(), df_gps['UTM_easting'].tolist(), label='obtained')
plt.plot(x, best_line(x), label='ideal')
plt.xlabel('UTM_northing')
plt.ylabel('UTM_easting')
plt.legend()
plt.title('ideal path vs obtained path')
```

```python
rsme = np.sqrt(np.square(y - y_ideal).mean())
rsme
```
